```python
import pandas as pd
```

```python
mushroomAbbv_df = pd.read_csv('mushrooms_abbreviated.csv')
mushroomAbbv_df.head()
```

```python
mushroom_df = mushroomAbbv_df.copy()
```

```python
mushroom_df['class'] = mushroom_df['class'].map({'p':'poisonous', 'e':'edible'})
mushroom_df['cap-shape'] = mushroom_df['cap-shape'].map({
    'b':'bell', 'c':'conical', 'x':'convex', 'f':'flat', 'k':'knobbed', 's':'sunken'})
mushroom_df['cap-surface'] = mushroom_df['cap-surface'].map({'f':'fibrous', 'g':'grooves', 'y':'scaly', 's':'smooth'})
mushroom_df['cap-color'] = mushroom_df['cap-color'].map({
    'n':'brown', 'b':'buff', 'c':'cinnamon', 'g':'gray', 'r':'green', 'p':'pink', 'u':'purple', 'e':'red',
    'w':'white', 'y':'yellow'})
mushroom_df['bruises'] = mushroom_df['bruises'].map({'t':'bruises', 'f':'no_bruises'})
mushroom_df['odor'] = mushroom_df['odor'].map({
    'a':'almond', 'l':'anise', 'c':'creosote', 'y':'fishy', 'f':'foul', 'm':'musty', 'n':'none',
    'p':'pungent', 's':'spicy'})
mushroom_df['gill-attachment'] = mushroom_df['gill-attachment'].map({
    'a':'attached', 'd':'descending', 'f':'free', 'n':'notched'})
mushroom_df['gill-spacing'] = mushroom_df['gill-spacing'].map({'c':'close', 'w':'crowded', 'd':'distant'})
mushroom_df['gill-size'] = mushroom_df['gill-size'].map({'b':'broad', 'n':'narrow'})
mushroom_df['gill-color'] = mushroom_df['gill-color'].map({
    'k':'black', 'n':'brown', 'b':'buff', 'h':'chocolate', 'g':'gray', 'r':'green', 'o':'orange', 'p':'pink',
    'u':'purple', 'e':'red', 'w':'white', 'y':'yellow'})
mushroom_df['stalk-shape'] = mushroom_df['stalk-shape'].map({'e':'enlarging', 't':'tapering'})
mushroom_df['stalk-root'] = mushroom_df['stalk-root'].map({
    'b':'bulbous', 'c':'club', 'u':'cup', 'e':'equal', 'z':'rhizomorphs', 'r':'rooted', '?':'missing'})
mushroom_df['stalk-surface-above-ring'] = mushroom_df['stalk-surface-above-ring'].map({
    'f':'fibrous', 'y':'scaly', 'k':'silky', 's':'smooth'})
mushroom_df['stalk-surface-below-ring'] = mushroom_df['stalk-surface-below-ring'].map({
    'f':'fibrous', 'y':'scaly', 'k':'silky', 's':'smooth'})
mushroom_df['stalk-color-above-ring'] = mushroom_df['stalk-color-above-ring'].map({
    'n':'brown', 'b':'buff', 'c':'cinnamon', 'g':'gray', 'o':'orange', 'p':'pink', 'e':'red',
    'w':'white', 'y':'yellow'})
mushroom_df['stalk-color-below-ring'] = mushroom_df['stalk-color-below-ring'].map({
    'n':'brown', 'b':'buff', 'c':'cinnamon', 'g':'gray', 'o':'orange', 'p':'pink', 'e':'red',
    'w':'white', 'y':'yellow'})
mushroom_df['veil-type'] = mushroom_df['veil-type'].map({'p':'partial', 'u':'universal'})
mushroom_df['veil-color'] = mushroom_df['veil-color'].map({'n':'brown', 'o':'orange', 'w':'white', 'y':'yellow'})
mushroom_df['ring-number'] = mushroom_df['ring-number'].map({'n':'none', 'o':'one', 't':'two'})
mushroom_df['ring-type'] = mushroom_df['ring-type'].map({
    'c':'cobwebby', 'e':'evanescent', 'f':'flaring', 'l':'large', 'n':'none', 'p':'pendant',
    's':'sheathing', 'z':'zone'})
mushroom_df['spore-print-color'] = mushroom_df['spore-print-color'].map({
    'k':'black', 'n':'brown', 'b':'buff', 'h':'chocolate', 'r':'green', 'o':'orange', 'u':'purple',
    'w':'white', 'y':'yellow'})
mushroom_df['population'] = mushroom_df['population'].map({
    'a':'abundant', 'c':'clustered', 'n':'numerous', 's':'scattered', 'v':'several', 'y':'solitary'})
mushroom_df['habitat'] = mushroom_df['habitat'].map({
    'g':'grasses', 'l':'leaves', 'm':'meadows', 'p':'paths', 'u':'urban', 'w':'waste', 'd':'woods'})

mushroom_df.head()
```

```python
mushroom_df.to_csv('mushrooms_expanded.csv', index=False)
```
